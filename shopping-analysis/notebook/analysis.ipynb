```python
df['final_price'] = (df['final_price']
    .astype(str)
    .str.replace('₹','',regex=False)
    .str.replace(',','',regex=False)               
    .str.replace('"','',regex=False)                
)
df['final_price'] = pd.to_numeric(df['final_price'],errors='coerce')                   
```

```python
df[['initial_price','final_price','discount']].dtypes
```

```python
df['discount'] = df['discount'].fillna(0)
df['what_customers_said'] = df['what_customers_said'].fillna('No Review')
df['seller_name'] = df['seller_name'].fillna('Unknown')
df['videos'] = df['videos'].fillna('No Video')
df['seller_information'] = df['seller_information'].fillna('Not Available')
df['variations'] = df['variations'].fillna('No Variation')
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df['price_difference'] = df['initial_price'] - df['final_price']
```

```python
df[['title','price_difference']].head()
```

```python
df['popularity'] = df['rating'] * df['ratings_count']
```

```python
df[['title','popularity']].head()
```

```python
df['rating'].describe()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['rating'], bins=10)
plt.title("Distribution of Ratings")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['rating'])
plt.title("Boxplot of Ratings")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='ratings_count', y='rating', data=df)
plt.title("Rating vs Ratings Count")
plt.show()
```

```python
category_rating = df.groupby('category')['rating'].mean().sort_values(ascending=False)
category_rating.head(10)
```

```python
plt.figure(figsize=(10,5))
category_rating.head(10).plot(kind='bar')
plt.title("Top Categories by Average Rating")
plt.show()
```

# Insights
- Most products have ratings between 4 and 5, indicating generally positive customer feedback.
- Some categories have higher average ratings than others, showing better customer satisfaction in those categories.
- Products with a higher number of ratings tend to be more popular.
- Missing values were handled and duplicate records were removed to improve data quality.
- Feature engineering was performed by creating price_difference and popularity columns for better analysis.

# Business Implications
- Highly rated products tend to receive more customer engagement.
- Categories with high popularity metrics can be prioritized in promotions.
- Monitoring price differences helps evaluate discount effectiveness.
- Popular products can be recommended to improve sales and customer satisfaction.
